```python
import numpy as np
from scipy.integrate import solve_ivp

%matplotlib agg
import matplotlib as mpl

import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib import rc

from IPython.display import HTML
rc("animation", html="html5")
rc("font", size=12)
```

```python
g = 10.0
l = 1.0

def pendulum(t, z, u):
    theta = z[0]
    omega = z[1]
    
    return np.array([omega, -g / l * np.sin(theta) + u(t) * 1.0 / l * np.sin(theta)])

sol0 = solve_ivp(lambda t, z: pendulum(t, z, lambda t: 0.0), [0, 20.0], [0.99 * np.pi, 0.0], max_step=0.01)
hmax = 0.2
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(5, 5), dpi=150)
ax.plot(sol0.t, sol0.y[0, :])

plt.show()
plt.close(fig)
```

```python
%matplotlib agg
fig, ax = plt.subplots(figsize=(5, 5), dpi=150)


scat = ax.scatter(l * np.sin(sol0.y[0, 0]), -l * np.cos(sol0.y[0, 0]), s=100, antialiased=True)
scat2 = ax.scatter(0.0, 0.0, s=10, c="r", zorder=10)
line, = ax.plot([l * np.sin(sol0.y[0, 0]), 0.0], [-l * np.cos(sol0.y[0, 0]), 0.0], c="k", zorder=-10)
ax.plot([0.0, 0.0], [0.0, hmax], c="k", lw=0.5)
ax.plot([0.0, 0.0], [-0.025, hmax + 0.025], c="k", lw=6.0, alpha=0.5)

ax.set_xlim([-l - 0.1, l + 0.1])
ax.set_ylim([-l - 0.1, l + 0.1])

def update(frame_idx):
    scat.set_offsets(np.array([[l * np.sin(sol0.y[0, frame_idx])], [-l * np.cos(sol0.y[0, frame_idx])]]).T)
    line.set_xdata([l * np.sin(sol0.y[0, frame_idx]), 0.0])
    line.set_ydata([-l * np.cos(sol0.y[0, frame_idx]), 0.0])
    
    return [scat, line]

anim = animation.FuncAnimation(fig, update, frames=685, interval=15, blit=True)
plt.close(fig)
anim
```

```python
f = 10
hmax = 0.2
h = lambda t: hmax * (1 + np.sin(2 * np.pi * f * t)) / 2
u = lambda t: -hmax * (2 * np.pi * f) ** 2 * np.sin(2 * np.pi * f * t) / 2

sol = solve_ivp(lambda t, z: pendulum(t, z, u), [0, 20.0], [0.99 * np.pi, 0.0], max_step=0.01)
```

```python
%matplotlib agg
fig, ax = plt.subplots(figsize=(5, 5), dpi=150)

scat = ax.scatter(l * np.sin(sol.y[0, 0]), -l * np.cos(sol.y[0, 0]), s=100)
scat2 = ax.scatter(0.0, 0.0, s=10, c="r", zorder=10)
plot, = ax.plot([l * np.sin(sol.y[0, 0]), 0.0], [-l * np.cos(sol.y[0, 0]), 0.0], c="k", zorder=-10)
ax.plot([0.0, 0.0], [0.0, hmax], c="k", lw=0.5)
ax.plot([0.0, 0.0], [-0.05, hmax + 0.05], c="k", lw=5.0, alpha=0.5)

ax.set_xlim([-(l + hmax) - 0.1, (l + hmax) + 0.1])
ax.set_ylim([-(l + hmax) - 0.1, (l + hmax) + 0.1])

def update(frame_idx):
    scat.set_offsets(np.array([[l * np.sin(sol.y[0, frame_idx])], [-l * np.cos(sol.y[0, frame_idx]) + h(sol.t[frame_idx])]]).T)
    scat2.set_offsets(np.array([[0.0], [h(sol.t[frame_idx])]]).T)

    plot.set_xdata([l * np.sin(sol.y[0, frame_idx]), 0.0])
    plot.set_ydata([-l * np.cos(sol.y[0, frame_idx]) + h(sol.t[frame_idx]), h(sol.t[frame_idx])])
    
    return [scat, scat2, plot]

anim = animation.FuncAnimation(fig, update, frames=200, interval=15)
plt.close(fig)
anim
```
