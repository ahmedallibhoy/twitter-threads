```python
import numpy as np

from scipy.integrate import solve_ivp, simpson
from scipy.interpolate import interp1d

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib import rc

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})

import os

#from IPython.display import HTML
#rc("animation", html="html5")
#rc("font", size=12)
```

```python
l = 1.0
g = 9.8
hmax = 1.1 * np.sqrt(2 * g)
beta = 0.01
```

```python
def pendulum(t, x, u):
    theta, p = x
    dtheta = p
    dp = -beta / (l ** 2) * p + g / l * np.sin(theta) + u(t) / l * np.sin(theta)
    return np.array([dtheta, dp])

def pendulum_unforced(t, x, b):
    theta, p = x
    dtheta = p
    dp = -b * p + g / l * np.sin(theta)
    return np.array([dtheta, dp])

def pendulum_lin(t, x, u):
    theta, p = x
    dtheta = p
    dp = -beta / (l ** 2) * p + g / l * theta + u(t) / l * theta
    return np.array([dtheta, dp])

def pendulum_lin_av(t, x, A_av):
    return A_av @ x
```

```python
def get_unforced_sol(beta, Tmax, x0=np.array([0.01, 0])):
    return solve_ivp(lambda t, x: pendulum_unforced(t, x, beta), [0, Tmax], x0, max_step=Tmax / 2000)
```

```python
sol_unforced = get_unforced_sol(0.5, 10)
theta_range = np.linspace(-1.0, 1.05 * np.max(sol_unforced.y[0, :]), 20)
p_range = np.linspace(-10, 5.5 , 20)

Theta, P = np.meshgrid(theta_range, p_range)
dTheta = P
dP = -0.5 * P + g / l * np.sin(Theta)
dN1 = dTheta / (1e-5 + np.sqrt(dTheta ** 2 + dP ** 2))
dN2 = dP / (1e-5 + np.sqrt(dTheta ** 2 + dP ** 2))

t_range = np.linspace(0, sol_unforced.t[-1], 200)
x_interp = interp1d(sol_unforced.t, sol_unforced.y)(t_range)

l0 = 0.5
h0 = 0.025

for idx, t in enumerate(t_range):
    fig, ax = plt.subplots(figsize=(6, 5), dpi=150)

    ax.quiver(Theta, P, dN1, dN2, angles="xy")
    ax.plot(sol_unforced.y[0, :], sol_unforced.y[1, :])
    ax.scatter(0, 0, s=50, marker="x", color="r", zorder=100)
    ax.scatter(x_interp[0, idx], x_interp[1, idx], s=20, zorder=101)
    
    ax.set_xlabel(r"$\theta$")
    ax.set_ylabel(r"$\dot{\theta}$")

    inset_ax = fig.add_axes([0.15, 0.15, 0.35, 0.35], facecolor='w')
    #inset_ax.patch.set_alpha(1.0)
    inset_ax.scatter(l0 * np.sin(x_interp[0, idx]), l0 * np.cos(x_interp[0, idx]), s=100, antialiased=True)
    inset_ax.scatter(0.0, 0.0, s=10, c="r", zorder=10)
    inset_ax.plot([l0 * np.sin(x_interp[0, idx]), 0.0], [l0 * np.cos(x_interp[0, idx]), 0.0], c="k", zorder=-10)
    
    inset_ax.xaxis.set_tick_params(labelbottom=False)
    inset_ax.yaxis.set_tick_params(labelleft=False)

    inset_ax.set_xlim([-l0 - 0.05, l0 + 0.05])
    inset_ax.set_ylim([-l0 - h0 - 0.05, l0 + h0 + 0.05])

    plt.savefig("unforced-phase/%04d.png" % idx, transparent=False, bbox_inches="tight")
    #plt.show()
    plt.close(fig)
    
os.popen("convert -background white -alpha remove -layers OptimizePlus -delay 1x30 -dispose previous -loop 0 unforced-phase/*.png unforced-phase.gif")
```

```python
A_av = np.array([[0, 1], [-hmax ** 2 / (2 * l ** 2) + g / l, -beta / (l ** 2)]])
o_av = np.imag(np.linalg.eigvals(A_av))[0]
tau = 2 * np.pi / o_av

def get_sols(eps, Tmax, x0):
    omega = 1.0 / eps
    h = lambda t: (hmax / omega) * np.sin(omega * t)
    hdot = lambda t: hmax * np.cos(omega * t)
    u = lambda t: -(omega) * hmax * np.sin(omega * t)

    sol = solve_ivp(lambda t, x: pendulum(t, x, u), [0, Tmax], x0, max_step=Tmax / 2000)
    sol_lin = solve_ivp(lambda t, x: pendulum_lin(t, x, u), [0, Tmax], x0, max_step=Tmax / 2000)

    t_range = np.arange(0, 2 * np.pi / omega, Tmax / 2000)
    y_interp = interp1d(sol.t, sol.y)(t_range)
    x0_av = omega / (2 * np.pi) * simpson(y_interp, t_range)
    sol_lin_av = solve_ivp(lambda t, x: pendulum_lin_av(t, x, A_av), [0, Tmax], x0_av, max_step=Tmax / 10000)
    
    return sol, sol_lin, sol_lin_av
```

```python
eps = 0.015
x0 = np.array([0.025, 0]) 
Tmax = 6 * tau
sol, sol_lin, sol_lin_av = get_sols(eps, Tmax, x0)
```

```python
%matplotlib inline

l0 = 0.5
h0 = 0.025
hh = lambda t: h0 * np.sin(1.0 / eps * t)

t_range = np.linspace(0, 2.0 * tau, 60)
x_interp = interp1d(sol.t, sol.y)(t_range)

for idx, t in enumerate(t_range):
    fig, ax = plt.subplots(figsize=(5, 5), dpi=150)

    scat = ax.scatter(l0 * np.sin(x_interp[0, idx]), hh(t) + l0 * np.cos(x_interp[0, idx]), s=100, antialiased=True)
    scat2 = ax.scatter(0.0, hh(t), s=10, c="r", zorder=10)
    line, = ax.plot([l0 * np.sin(x_interp[0, idx]), 0.0], [hh(t) + l0 * np.cos(x_interp[0, idx]), hh(t)], c="k", zorder=-10)

    ax.plot([0.0, 0.0], [-h0, h0], c="k", lw=0.5)
    ax.plot([0.0, 0.0], [-0.025 - h0, h0 + 0.025], c="k", lw=6.0, alpha=0.5)

    ax.set_xlim([-l0 - 0.05, l0 + 0.05])
    ax.set_ylim([-l0 - h0 - 0.05, l0 + h0 + 0.05])
    
    plt.axis("off")
    plt.savefig("forced/%04d.png" % idx, transparent=True)
    plt.close(fig)
    
os.popen("convert -background white -alpha remove -layers OptimizePlus -delay 1x30 -dispose previous -loop 0 forced/*.png pendulum-forced.gif")
```
